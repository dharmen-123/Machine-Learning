## ********** Types of Machine Learning ***********

```python
# Machine Learning is mainly divided into 4 types based on the kind of data available and how the model learns.

# Machine Learning
# │
# ├── 1. Supervised Learning
# ├── 2. Unsupervised Learning
# ├── 3. Semi-Supervised Learning
# └── 4. Reinforcement Learning
```

#### 1. Supervised Learning
Definition

In Supervised Learning, the model learns from labeled data.

Labeled data means the correct answer (output) is already known.

Formula
Input + Correct Output → Model Learns

Real-Life Examples

* House Price Prediction
* Spam Email Detection
* Loan Approval
* Disease Prediction
* Student Result Prediction

Types

* Regression → Predicts continuous values (e.g., house price, salary).
* Classification → Predicts categories (e.g., spam/not spam, pass/fail).

#### 2. Unsupervised Learning
Definition

In Unsupervised Learning, the model learns from unlabeled data.

There are no correct answers provided. The model finds hidden patterns or groups on its own.

Formula
Input Data → Find Patterns

Real-Life Examples
 * Customer Segmentation
 * Product Recommendation
 * Market Basket Analysis
 * Fraud Pattern Detection

Types

* Clustering
* Association Rule Learning
* Dimensionality Reduction

#### 3. Semi-Supervised Learning
Definition

Semi-Supervised Learning uses a small amount of labeled data and a large amount of unlabeled data.

It combines ideas from supervised and unsupervised learning.

Formula
Few Labeled Data + Many Unlabeled Data → Better Model

Real-Life Examples

* Medical Image Classification
* Speech Recognition
* Face Recognition
* Document Classification

#### 4. Reinforcement Learning
Definition

* In Reinforcement Learning, an agent learns by interacting with an environment.

The agent receives:
 * Reward for a good action.
 * Penalty for a bad action.

Over time, it learns the best strategy.

Formula
Action → Reward/Penalty → Learn → Better Action

Real-Life Examples

* Self-Driving Cars
* Robotics
* Game Playing (Chess, Go)
* Traffic Signal Optimization
* Robot Navigation

####  Interview Answers

* Supervised Learning: Learns from labeled data to predict outputs.
* Unsupervised Learning: Learns from unlabeled data to discover hidden patterns.
* Semi-Supervised Learning: Learns using a small amount of labeled data and a large amount of unlabeled data.
* Reinforcement Learning: Learns the best actions by interacting with an environment and receiving rewards or penalties.
