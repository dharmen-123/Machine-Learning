## *************** Machine Learning ***************

#### What is Machine learning ?

Machine learning is a branch of Artificial Intelligence that focuses on developing models and algorithms.

ML is the way of teaching computers to learns from data-just like we human learns from experience


#### Why Do We Need Machine Learning?

Traditional programming works well when the rules are clear and fixed. However, many real-world problems are too complex to solve with manually written rules.

Machine Learning helps computers:

Learn from data automatically.
Improve performance with more data.
Make predictions about future events.
Recognize patterns that humans may miss.
Handle large and complex datasets efficiently.

#### Traditional Programming vs Machine Learning

The main difference between Traditional Programming and Machine Learning is how the computer produces the output.

In Traditional Programming, the programmer writes all the rules.

Workflow.
    
    Input Data + Program (Rules) → Output

In Machine Learning, the computer learns the rules from data.

Workflow.

Training Phase

    Input Data + Correct Output → Machine Learning Algorithm → Trained Model

Prediction Phase

    New Input → Trained Model → Predicted Output
