RFE ( Recursive Feature Elimination )

```python
# Step 1: Import libraries
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE

# Step 2: Load dataset
X, y = load_iris(return_X_y=True)

# Step 3: Define the model
model = LogisticRegression(max_iter=200)

# Step 4: Apply Recursive Feature Elimination (RFE)
# Select top 2 features
rfe = RFE(estimator=model, n_features_to_select=2)
fit = rfe.fit(X, y)

# Step 5: Results
print("Selected Features (True=Selected):", fit.support_)
print("Feature Ranking:", fit.ranking_)
```

```python
# Step 1: Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE

# Step 2: Load dataset
iris = load_iris()
X, y = iris.data, iris.target
feature_names = iris.feature_names

# Step 3: Define the model
model = LogisticRegression(max_iter=200)

# Step 4: Apply Recursive Feature Elimination (RFE)
# Select top 2 features
rfe = RFE(estimator=model, n_features_to_select=2)
fit = rfe.fit(X, y)

# Step 5: Print results
print("Selected Features (True=Selected):", fit.support_)
print("Feature Ranking:", fit.ranking_)

# Step 6: Plot feature rankings
plt.figure(figsize=(8,5))
plt.bar(feature_names, fit.ranking_, color='skyblue')
plt.xlabel("Features")
plt.ylabel("Ranking (1 = most important)")
plt.title("Feature Ranking using RFE (Wrapper Method)")
plt.show()
```
