#### ******************** Data Preprocessing ********************

Data Preprocessing is the process of preparing raw data into a clean and suitable format before training a Machine Learning model.

Simple Definition

Data Preprocessing is the process of cleaning and transforming raw data so that a machine learning model can use it effectively.

Why is Data Preprocessing Important?

Machine Learning models cannot work properly with raw data because it may contain:

* Missing values
* Duplicate records
* Incorrect data
* Categorical (text) values
* Different data scales

Data preprocessing solves these issues and improves model performance.

#### Real-Life Example

Imagine you are going to cook rice.

Before cooking, you:
* Wash the rice.
* Remove stones or impurities.
* Measure the correct amount of water.

Only then do you cook it.

Similarly, before training a machine learning model, we preprocess the data to make it ready for learning.

#### -------- Steps in Data Preprocessing ---------

1. Data Cleaning

Remove or fix missing values, duplicate records, incorrect data, and outliers.

Example:

Age = NaN  → Fill with mean

2. Data Transformation

Convert data into a format suitable for machine learning.

Example:

Male → 1
Female → 0

3. Feature Scaling

Bring numerical features to a similar scale.

Example:

Before:

Age = 25
Salary = 500000

After scaling:

Age = 0.35
Salary = 0.62
