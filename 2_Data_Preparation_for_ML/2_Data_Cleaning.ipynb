### ************** Data Cleaning **************

Data Cleaning is the process of finding and fixing errors, missing values, duplicate records, incorrect data, and inconsistent data so that the dataset becomes clean, accurate, and ready for analysis or machine learning.

Simple Definition

Data Cleaning is the process of making raw data clean, correct, complete, and consistent before using it for Machine Learning.

Why is Data Cleaning Important?

Machine Learning models learn from the data we provide.

* If the data is clean, the model performs well.
* If the data is dirty or incorrect, the model may produce inaccurate predictions.

Remember:

"Garbage In = Garbage Out (GIGO)"

This means if you give poor-quality data to a machine learning model, you will get poor-quality predictions.

#### Data Cleaning Workflow

* Raw Data
*     ↓
* Check Missing Values
*     ↓
* Remove Duplicates
*     ↓
* Correct Incorrect Data
*     ↓
* Handle Outliers
*     ↓
* Fix Inconsistent Data
*     ↓
* Convert Data Types
*     ↓
* Clean Data
*     ↓
* Ready for Machine Learning

```python
import pandas as pd
import numpy as np

data = {
    "Employee_ID": [101,102,103,104,105,106,106,107,108,109],
    "Name": ["Amit","Rahul","Neha","Priya","Karan","Riya","Riya","Mohit","Anjali","Vikas"],
    "Age": [25,np.nan,30,-5,28,35,35,29,150,27 ],
    "Salary": [ 40000, 50000, np.nan, 45000, 5000000, 55000, 55000, 48000, 47000, 46000 ],
    "Gender": ["Male","Female","female","Male","MALE","Female","Female","male","Female","Male"],
    "City": ["Delhi","delhi","DELHI","Mumbai","mumbai","Bhopal","Bhopal","bhopal","Indore","INDORE"],
    "Joining_Date": ["2021-01-10","2022/02/15","15-03-2021","2020-05-20","2021/07/18","2023-08-10","2023-08-10","2022-11-12","2021/01/05","2020-09-01"],
    "Phone_Number": ["9876543210","9876543211","9876543212","9876543213","9876543214","9876543215","9876543215","9876543216","9876543217","9876543218"]
}
df = pd.DataFrame(data)
print(df)
```

Step 1: Handle Missing Values

Missing values occur when some information is not available in the dataset. We either fill the missing values or remove them.

```python

# Check missing values
df.isnull().sum()

# Fill missing values
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Salary"].fillna(df["Salary"].median(), inplace=True)

# OR remove rows with missing values
df.dropna(inplace=True)
```

Step 2: Remove Duplicate Records

Duplicate records are repeated rows in the dataset. Removing them prevents biased analysis and improves model quality.

```python
# Check duplicate rows
df.duplicated().sum()

# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

Step 3: Correct Incorrect Data

Sometimes data is entered incorrectly, such as negative ages or impossible values. These should be corrected or removed.

```python
# Keep only valid ages
df = df[df["Age"] > 0]
df
```

Step 4: Handle Outliers

Outliers are values that are much larger or smaller than the rest of the data. They can affect model performance.

```python
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["Salary"] >= lower) & (df["Salary"] <= upper)]
```

Step 5: Fix Inconsistent Data

The same information may appear in different formats. Convert it to a consistent format.

```python
df.loc[:, "Gender"] = df["Gender"].str.capitalize()
```

Step 6: Convert Data Types

Ensure each column has the correct data type, such as integer, float, string, or date.

```python
df.loc[:, "Age"] = df["Age"].astype(int)
df.loc[:, "Salary"] = df["Salary"].astype(float)
```

Step 7: Remove Unnecessary Columns

Columns that do not help in prediction should be removed to simplify the dataset.

Step 8: Rename Column Names

Rename columns to make them more meaningful and easier to understand.

Step 9: Handle Categorical Data

Machine learning models cannot work directly with text values, so convert categorical values into numbers.

Step 10: Check Final Dataset

After cleaning, verify that the dataset is free of missing values, duplicates, and incorrect data before using it for machine learning.

```python
# Dataset information
df.info()

# Summary statistics
df.describe()

# Missing values
df.isnull().sum()

# Duplicate values
df.duplicated().sum()

# First five rows
df.head()
```
