### ***** Exploratory Data Analysis (EDA) *****

EDA (Exploratory Data Analysis) is the process of understanding, exploring, and analyzing a dataset before building a Machine Learning model.

Its main purpose is to answer questions like:

* What does the data look like?
* Is the data clean?
* Are there missing values?
* Are there outliers?
* How are the features related?
* Which features are important?

In simple words:

EDA means "getting to know your data before using it to train a machine learning model."

#### EDA Workflow

* Collect Data
*       ↓
* Understand Dataset
*       ↓
* Check Missing Values
*       ↓
* Remove Duplicates
*       ↓
* Handle Outliers
*       ↓
* Analyze Data Distribution
*       ↓
* Study Feature Relationships
*       ↓
* Visualize Data
*       ↓
* Prepare Data for Machine Learning

Steps in EDA

1. Viewing the Data
* head() , tail() , shape() , info() , dtypes

2. Summary Statistics
* mead , median  ,mode , std , min , max , quartiles
* Helps understand spread and central tendency

3. Values Count
* How many unique values in your coloumn ?
* Great for Categorical columns

4. Missing Values Analysis 
* Indentify the Gaps in data ?
* What percent of data is missing ?

5. Visualizations
* Histogram 
* boxplot
* countplot
* Scatter plot
* Correlation Heatmaps
* etc..

6. Target Variable Exploration 
* how does your output relate to other variables
